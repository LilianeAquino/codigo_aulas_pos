# Removendo tokens indesejados ou normalizando o texto

A remoção de tokens que não possuem informação para a sua tarefa, permite diminuir o modelo e a quantidade de informação armazenada. Além disso, não gera ruído quando usado por um algoritmo. Esse ruído pode atrapalhar uma classificação, por exemplo.

```python
# certifique-se que você tem o corpus tentando fazer o download. 
import nltk
nltk.download("machado")
```

```python
from nltk.corpus import machado
import string
```

```python
conto01 = machado.raw(machado.fileids()[0]) 

# podemos querer remover a pontuação se ela não for útil (em muitos casos ela realmente não é útil)

# montagem da tabela de pontuacoes, primeiro pegamos a lista de pontuacoes com string.punctuation, depois pegamos
# e criamos como uma tabela usando o str.maketrans
table = str.maketrans(dict.fromkeys(string.punctuation)) 
conto01 = conto01.translate(table)
```

```python
from gensim.utils import simple_preprocess
```

```python
tokens_conto01 = simple_preprocess(conto01)
```

```python
# Remover numeros, mas não palavras que contém números
tokens_conto01 = [token for token in tokens_conto01 if not token.isnumeric()]
```

```python
# veja que numeros romanos permaneceram, caso queira remove-los, é possível fazer uma expressão regular ou uma lista 
# simples para remover esses itens
print(tokens_conto01) 
```

```python
# Remove palavras que contém apenas um caracter. Por exemplos, artigos definidos, como "a" e "o"
# caso nomes próprios ou letras maísculas não sejam importantes, podemos ao mesmo tempo transformar o token com 
# lower
tokens_conto01 = [token.lower() for token in tokens_conto01 if len(token)>1]
```

# Remoção de stop-words

Stop words são palavras muito frequentes e que não acrescentam nada. Por exemplo, preposições e conjunções como "de". Existem stop words que podem ser ligadas ao domínio, neste caso você precisa se um especialista para determinar isso.

```python
import nltk
stopwords = nltk.corpus.stopwords.words('portuguese')
```

```python
tokens_conto01 = [token for token in tokens_conto01 if token not in stopwords]
```

```python
print(tokens_conto01)
```

# Stemming

A stemização de palavras é o processo de capturar a "raiz" da palavra

```python
stemmer = nltk.stem.RSLPStemmer() # esse stemmer eh para portugues
```

```python
tokens_conto01_stem = [stemmer.stem(token) for token in tokens_conto01]
```

# Lematização

Determina qual é forma de "dicionário" de uma palavra, i.e., retira a flexão dela. Se for um verbo, por exemplo, ele coloca no infinitivo ou se um substantivo está no plural, ele fica no singular. Assim essa tarefa não é tão simples e portanto nem sempre existem soluções satisfatórias para a língua que você está trabalhando.

Existem poucas ferramentas para a lematização no português. Temos o spacy, porém sua lematização é bem deficiente. Em Java existe o LemPORT, porém consome muita memória e processamento.

```python
import spacy
```

```python
nlp = spacy.load("pt_core_news_sm")
```

```python
doc_conto01 = nlp(conto01)
```

```python
for token in doc_conto01:
    print(token.lemma_)
```
